# System and Process Automation with Python

### Challenge:

For cost control, every day your boss asks for a report with all the company's purchases of goods.
Your job, as an analyst, is to send him an e-mail, as soon as you start work, with the total spent, the quantity of products bought, and the average price of the products.

Observations: This project was idealized by Hashtag training sessions in order to introduce python possibilities for beginners with applications for the real world. In this challenge I was able to regain my taste for automation and also add to the project with some tools that I felt were necessary such as formatting number in the brazilian currency using locale module and being able to find the most recent file in the folder with os and glob modules. Time sleep was extensively used in order to wait my computer to execute commands before moving to the next ones. Next step is to replace those repeated lines of code with commands that consider these delays.

Suggestions are welcome! Thanks for the attention in this simple project :)

```python
import pyautogui
import pandas as pd
import time
import os.path
import glob
import locale

locale.setlocale(locale.LC_ALL,'pt_BR.UTF-8')

# Step 1: Open the browser
pyautogui.press("win")
time.sleep(4)
pyautogui.write("chrome")
time.sleep(2)
pyautogui.press("enter")
time.sleep(4)

# Step 2: Open company system
pyautogui.click(x=-1110, y=50)
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")
time.sleep(10)

pyautogui.press("tab")
pyautogui.write("login")
pyautogui.press("tab")
pyautogui.write("senha")
pyautogui.press("tab")
pyautogui.press("enter")
time.sleep(15)

# Step 3: Download file
pyautogui.click(x=-114, y=375, button = "right")
time.sleep(5)
pyautogui.click(x=-408, y=581)
time.sleep(5)

# Step 4: Read downloaded file data
folder_path = r'C:\Users\guido\Downloads'
file_type = r'\*csv'
files = glob.glob(folder_path + file_type)
max_file = max(files, key=os.path.getctime)
table = pd.read_csv(max_file, sep = ";")

# Step 5: Calculate indicators such as total expenses, products quantity and products average price
gasto = table['ValorFinal'].sum()
quantidade = table['Quantidade'].sum()
precomedio = gasto / quantidade


# Step 6: Send email
pyautogui.hotkey("ctrl", "t") 

pyautogui.write("gmail.com")
pyautogui.press("enter")
time.sleep(15)

pyautogui.click(x=-1282, y=195)
time.sleep(10)
pyautogui.write("guidotognini@usp.br")
pyautogui.press("tab")
time.sleep(2)
pyautogui.press("tab")
pyautogui.write("Automated mail")
pyautogui.press("tab")
texto = f"""
Dear boss, 

I have no reason why spend precious time every day sending you this email when I can just automate it with python from the lessons I learned.

Check this out: 

Total Expenses: {locale.currency(gasto, grouping = True)}
Total Number of Products: {quantidade}
Average Price: {locale.currency(precomedio, grouping = True)}

Cheers
"""
time.sleep(5)
pyautogui.write(texto)
time.sleep(15)
pyautogui.hotkey("ctrl","enter")
time.sleep(10)
pyautogui.hotkey("alt","f4")
```
